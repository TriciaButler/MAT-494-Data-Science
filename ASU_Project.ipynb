```python
survey_data = pd.crosstab(index=df['Question4'], columns=df['Question5'], margins=True)
print(survey_data)

print("P(thinking of goals often|wants to be a leader): ", survey_data.iloc[2, 1] / survey_data.iloc[4, 1])
print("P(thinking of goals often|do not want to be a leader): ", survey_data.iloc[2, 0] / survey_data.iloc[4, 0])
```

```python
survey_data = pd.crosstab(index=df['Question2'], columns=df['Question3'], margins=True)
print(survey_data)

print("P(job|motivation is to improve future): ", survey_data.iloc[3, 0] / survey_data.iloc[6, 0])
print("P(college|motivation is to improve future): ", survey_data.iloc[2, 0] / survey_data.iloc[6, 0])
```
